# 2) Data Wrangling and cleaning

```python
data_map.isna().any()
```

```python
data_map.isna().sum()
```

There are several columns with NULL values which will impact the analysis. Let 's sort it out first

For the categorical features such as the sex and respondent, it is common that the respondent are neither landlord nor tenant, so we will specify them as 'other'. Same for sex, there is non binary gender.

=> respondent/sex/type_of_property/type_of_toilet/water_source :=> NaN will be replaced by 'other'


As for the numerical ones, they will be replaced by 0

```python
data_map.respondent.value_counts()
```

```python
data_map.sex.value_counts()
```

```python
categorical_features = ['respondent','sex','type_of_property','type_of_toilet','add_toilets','water_source']
numerical_features = ['plot_months','no_of_hhs','no_of_ppl','no_of_toilets','latitude','longitude','accuracy','altitude']
for c in categorical_features:
    if (c == 'water_source'):
        data_map[c] = data_map[c].replace(np.nan, 'other_specify')
    else:
        data_map[c] = data_map[c].replace(np.nan, 'Other')
        

for c in numerical_features:
    data_map[c] = data_map[c].replace(np.nan, 0)
```

```python
data_map.head()
```

```python
data_map.add_toilets.value_counts()
```

```python
cleaned_data = data_map[[                   1,                 'id',         'respondent',
                      'sex',        'plot_months',   'type_of_property',
                'no_of_hhs',          'no_of_ppl',           
           'type_of_toilet',      'no_of_toilets',        'add_toilets',
             'water_source',            'latitude',
                'longitude',           'accuracy',           'altitude']]
```

```python
cleaned_data.sample(20)
```

We can observe that there are unhabitated households. Such households are not needed for our analysis thus can be neglected.

```python
df_data = cleaned_data[cleaned_data['no_of_hhs']!=0]
```

```python
df_data.describe()
```

 From the cleaned data, we observe that median of the data are below 11 people (corresponding to 3 hhs) in a single plot and it's amazing the some plots contains 17 people in them with a total of 4 hhs.
 
 Ironically, the nb of toilets in those crowded households is just 1. 
 
 If we consider the plot_months as the duration of how long the plots have been habitated by the current hhs, then we can say that the 3rd Quartile hold the longest period with 144 months and just with 1 toilet shared by 4 hhs in it.
